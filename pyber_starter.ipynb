```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style="ticks",color_codes=True)

```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
file_1 = pd.read_csv(city_data_to_load)
file_1.head()
```

```python
ride_data_to_load = "data/ride_data.csv"
file_2 = pd.read_csv(ride_data_to_load)
file_2.head()

```

```python
merge_file = pd.merge(file_1, file_2, on='city', how='inner')

merge_file.head()
```

```python
merge_file.describe(include='object')
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python

merge_file = pd.DataFrame(np.random.rand(2375, 4), columns=['type', 'city','fare', 'driver_count'])
merge_file = np.random.RandomState(0)
x = merge_file.randn(100)
y = merge_file.randn(100)
colors = merge_file.rand(100)
sizes = 500 * merge_file.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.4,
            cmap='viridis')
plt.xlabel("City Type")
plt.ylabel("Fare")
plt.title("Bubble Plot of Ride Sharing Data")
 

plt.colorbar()
plt.tight_layout()
plt.savefig("../Images/fares_citytype.png")
plt.show()
```
